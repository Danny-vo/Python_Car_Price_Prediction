The model's forecasting performance will be evaluated using the Mean Absolute Percentage Error (MAPE). MAPE is a metric that measures the accuracy of a forecasting model by calculating the average absolute percentage difference between the predicted and actual values. It shows how far the predictions deviate from the actual values, expressed as a percentage. A lower MAPE indicates higher accuracy, meaning the model's predictions are closer to the actual data. Thus, the straightforward metric makes it simple to understand and compare the performance of different models. The Mean Absolute Percentage Error (MAPE) is defined mathematically as:

$$
\text{MAPE} = \frac{1}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right| \times 100
$$

Where:
- $n$ is the number of observations.
- $y_i$ is the actual value.
- $\hat{y}_i$ is the predicted value.

**Task 2, Question 1**: Clean **all** numerical features so that they can be used in training algorithms. For instance, back_legroom feature is in object format containing both numerical values and text. Extract numerical values (equivalently eliminate the text) so that the numerical values can be used as a regular feature.  
(2 marks)

```python
## Task 2, Question 1 Code Here

#import libraries
import pandas as pd
import numpy as np

#Handle invalid data
df.replace(to_replace='--', value=np.nan, inplace=True)
df.replace(to_replace='-', value=np.nan, inplace=True)
df.replace(to_replace='', value=np.nan, inplace=True)
df.replace(to_replace='UNKNOWN', value=np.nan, inplace=True)
df.replace(to_replace='None', value=np.nan, inplace=True)
df.replace(to_replace='Unspecified', value=np.nan, inplace=True)

#Right trim function returns cell value after trimming n characters right to left
def righttrim(x, n):
    if pd.isna(x): return x
    if isinstance(x,float): return x
    else: return str(x)[:(len(str(x))-n)]
    
#Left trim function returns cell value after trimming n characters left to right
def lefttrim(x, n):
    if pd.isna(x): return x
    if isinstance(x,float): return x
    else: return str(x)[n:]
    
#Process back_legroom
df['back_legroom']=df['back_legroom'].apply(lambda x: righttrim(x,3)).astype(float)

#Process front_legroom
df['front_legroom']=df['front_legroom'].apply(lambda x: righttrim(x,3)).astype(float)

#Process height
df['height']=df['height'].apply(lambda x: righttrim(x,3)).astype(float)

#Process maximum_seating
df['maximum_seating']=df['maximum_seating'].apply(lambda x: righttrim(x,6)).astype(float)

#Process wheelbase
df['wheelbase']=df['wheelbase'].apply(lambda x: righttrim(x,3)).astype(float)

#Process width
df['width']=df['width'].apply(lambda x: righttrim(x,3)).astype(float)
```

Here are the the steps I have taken in this section:

1.  I have identified different forms of NULL values in the dataset, which include '--","-","","UNKNOWN","None","Unspecified" and replace it with numpy value NULL

2. I convert all unclear nominal data into approriate numeric values by ommiting their units (Eg: 5 cm -> 5)

```python
#Splitting Power into maximum power and power rpm
power_rpm=df['power'].apply(lambda x: lefttrim(x,(str(x).find('@')+2))).apply(lambda x: righttrim(x,4)).apply(lambda x: str(x).replace(',','')).astype(float) #hp
max_power=df['power'].apply(lambda x: righttrim(x,(len(str(x))-str(x).find('@')+3))).astype(float) #RPM
df['max_power']=max_power
df['power_rpm']=power_rpm

#Splitting torque into maximum torque and torque rpm
torque_rpm=df['torque'].apply(lambda x: lefttrim(x,(str(x).find('@')+2))).apply(lambda x: righttrim(x,4)).apply(lambda x: str(x).replace(',','')).astype(float) #lb-ft 
max_torque=df['torque'].apply(lambda x: righttrim(x,(len(str(x))-str(x).find('@')+7))).astype(float) #RPM
df['max_torque']=max_torque
df['torque_rpm']=torque_rpm

#Creating variable called days_listed indicating time from listed_date to now
from datetime import date
df['listed_date']=pd.to_datetime(df['listed_date'])
df['days_listed'] = df['listed_date'].apply(lambda x: (pd.Timestamp(date.today()) - x).days)

#Creating variable called year_from_now indicating time from year to now
from datetime import date
df['year_from_now'] = df['year'].apply(lambda x: date.today().year - x)

#drop columns that are splitted and add columns that are created
df.drop(["power","torque","listed_date","year"], axis='columns', inplace=True)

numeric_var.remove("power")
numeric_var.append("max_power")
numeric_var.append("power_rpm")

numeric_var.remove("torque")
numeric_var.append("max_torque")
numeric_var.append("torque_rpm")

ordinal_var.remove("listed_date")
numeric_var.append("days_listed")

ordinal_var.remove("year")
numeric_var.append("year_from_now")
```

```python
#clean data
exclusion=["vin","exterior_color","interior_color"]

##process engine_type
df['engine_type']=df['engine_type'].apply(lambda x: x[:2])

#choose 5 most common inout in each nominal columns
for var in nominal_var:
    if var not in exclusion:
        if df[var].nunique()>6:
            most_frequent_value=df[var].value_counts().nlargest(5)
            df[var] = df[var].apply(lambda x: x if x in most_frequent_value.index else 'other')
    
##process exterior color
#find most common color
from collections import Counter

#Normalize input format (all lowercase)
df['exterior_color'] = df['exterior_color'].str.lower()

#create a list containing all input words
exterior_split = df['exterior_color'].to_string().split() 
  
# Pass the split_it list to instance of Counter class. 
exterior_counter = Counter(exterior_split) 

# input values and their respective counts. 
exterior_most_occur = exterior_counter.most_common(20) 
  
print(exterior_most_occur) #('metallic', 3442), ('white', 2580), ('black', 2559), ('silver', 1458), ('blue', 1064)

#Create new column called main_exterior_color
df["main_exterior_color"]={}

#Create an array for 5 most common color
common_exterior_color=[exterior_most_occur[x][0] for x in range(5)]

#Inputing data into main_exterior_color
for i in range (0,len(df)):
    test=0
    word_include=df['exterior_color'][i].split()
    for j in range (0,len(common_exterior_color)):
        if common_exterior_color[j] in word_include:
            df["main_exterior_color"][i]=common_exterior_color[j]
            test=1
            continue
    if test==0: df["main_exterior_color"][i]="other"

##process interior_color
#find most common color
from collections import Counter

#Normalize input format (all lowercase)
df['interior_color'] = df['interior_color'].str.lower()
df['interior_color'] = df['interior_color'].str.replace("("," ")
df['interior_color'] = df['interior_color'].str.replace(")"," ")
df['interior_color'] = df['interior_color'].str.replace("/"," ")

#create a list containing all input words
interior_split = df['interior_color'].to_string().split() 
  
# Pass the split_it list to instance of Counter class. 
interior_counter = Counter(interior_split) 
print(interior_counter)  

#Crate a list of most common color based on interior_counter
interior_most_occur=["black","gray","brown","white","red"]

#Create new column called main_interior_color
df["main_interior_color"]={}

#Create a new exterior color column

for i in range (0,len(df)):
    test=0
    word_include=df['interior_color'][i].split()
    for j in range (0,len(interior_most_occur)):
        if interior_most_occur[j] in word_include:
            df["main_interior_color"][i]=interior_most_occur[j]
            test=1
            continue
    if test==0: df["main_interior_color"][i]="other"

#drop exterior_color and interior_color columns
df.drop(["exterior_color","interior_color"], axis='columns', inplace=True)

nominal_var.remove("exterior_color")
nominal_var.append("main_exterior_color")

nominal_var.remove("interior_color")
nominal_var.append("main_interior_color")
```

Linear Regression = $\hat{y}  = \beta_0 + \beta_1 x + \beta_2 x_2 +... + \beta_n x_n + \epsilon$

$\hat{y}$: dependent variables

$\beta_1, \beta_2, \beta_n$: weight coefficient of the explanatory variable


$x_1, x_2, x_n$: independent variables

$\epsilon$: random error

**2.2: Gradient Boosting**

First, we calculate the initial estimate, which is average of all points. Next, we calculate the residuals, which is the difference between all the estimates and the actual points. Then we train a model that has high bias (such as decision stump) on the residuals to predict those residuals. By doing this, we fill the gap between our initial estimate and our data. By adding the result, we get closer to our target function. We repeat these steps

**2.3: Random Forest**
Random Forest combines multiple decision trees to create a “forest.” Decision trees uses a flowchart like a tree structure to show the predictions that result from a series of feature-based splits. It starts with a root node and ends with a decision made by leaves.

<img src="rf.png" alt="Drawing" style="width: 800px;"/>


The logic behind the Random Forest model is that multiple uncorrelated models (the individual decision trees) perform much better as a group than they do alone. When using Random Forest for regression, the forest picks the average of the outputs of all trees. There is low (or no) correlation between the individual models—that is, between the decision trees that make up the larger Random Forest model. While individual decision trees may produce errors, the majority of the group will be correct, thus moving the overall outcome in the right direction.

```python
#3. Train the models
#3.1: Linear Regression

#tune hyperparameters via cross validation: there is no hyperparameters for Linear Regression to be tuned

#split the dataset
from sklearn.model_selection import train_test_split
X= df_train[['max_power', 'mileage']]
y= df_train['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=1)

#train the model
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
slr.fit(X_train, y_train)
y_pred = slr.predict(X_test)

#predictive performance: MAPE value
mape = mean_absolute_percentage_error(y_test, y_pred) * 100
print(f'MAPE: {mape:.2f}%')
```

```python
#3.2: Gradient Boosting

#split the dataset
X= df_train[['max_power', 'mileage']]
y= df_train['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=1)

#tune hyperparameters via cross validation
#create mape function to calculate Mean Absolute Percentage Error
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# fit a Gradient Boosting
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

gb = GradientBoostingRegressor(random_state=1)

#create parameters grid
params = {'n_estimators': range(100,1100,50)}

#mape scoring function
mape_scorer = make_scorer(mape, greater_is_better=False)

# initialize the Gradient Boosting
gb_cv = GridSearchCV(gb,
                     param_grid = params,
                     scoring=mape_scorer,
                     cv=5)

gb_cv.fit(X_train, y_train)

# Predict on the validation or test set with the best estimator
y_pred = gb_cv.best_estimator_.predict(X_train)

# Calculate MAPE using the custom function
best_mape_score = mape(y_train, y_pred)

# Print best parameters and MAPE score
print(f'Best parameters: {gb_cv.best_params_}')
print(f'MAPE with best parameters: {best_mape_score:.3f}')
```

```python
#3.3: Random Forest
#split the dataset
X= df_train[['max_power', 'mileage']]
y= df_train['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=1)

#tune hyperparameters via cross validation
#create mape function to calculate Mean Absolute Percentage Error
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# fit a Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rf = RandomForestRegressor()

#create parameters grid
params = {'n_estimators': range(100,1100,50)}

#mape scoring function
mape_scorer = make_scorer(mape, greater_is_better=False)

# initialize the Gradient Boosting
rf_cv = GridSearchCV(rf,
                     param_grid = params,
                     scoring=mape_scorer,
                     cv=5)

rf_cv.fit(X_train, y_train)

# Predict on the validation or test set with the best estimator
y_pred = rf_cv.best_estimator_.predict(X_train)

# Calculate MAPE using the custom function
best_mape_score = mape(y_train, y_pred)

# Print best parameters and MAPE score
print(f'Best parameters: {rf_cv.best_params_}')
print(f'MAPE with best parameters: {best_mape_score:.3f}')
```

```python
#download libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_percentage_error, make_scorer


#create mape function to calculate Mean Absolute Percentage Error
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100


#random forest
rf = RandomForestRegressor(random_state=1)

#create parameters grid
params = {'n_estimators': [700,1000]}  

#mape scoring function
mape_scorer = make_scorer(mape, greater_is_better=False)

# initialize the Random Forest Regressor
rf_cv = GridSearchCV(rf,
                     param_grid = params,
                     scoring=mape_scorer,
                     cv=5)

rf_cv.fit(X_train, y_train)

# Predict on the validation or test set with the best estimator
y_pred = rf_cv.best_estimator_.predict(X_test)

# Calculate MAPE using the custom function
best_mape_score = mape(y_test, y_pred)

# Print best parameters and MAPE score
print(f'Best parameters: {rf_cv.best_params_}')
print(f'MAPE with best parameters: {best_mape_score:.3f}')
```
